```python
cap=cv2.VideoCapture(0)
print(cap.isOpened())
while 1:
    ret,frame=cap.read()
    print(ret)
    print(cap.get(3))
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    plt.imshow(frame)
    if cv2.waitKey(1):
        break
cap.release()
```

```python
cap = cv2.VideoCapture(0)
cap.isOpened()
while(cap.isOpened()):
    ret, frame = cap.read()
    print(ret)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame',gray)
    if cv2.waitKey(1):
        break

cap.release()
cv2.destroyAllWindows()
```
