```python
def sample_type(subject_name):
    powder_list=['框架','粉末','矿物','菌']
    type_of_sample=""
    for key_word in powder_list:
        if key_word in subject_name:
            type_of_sample="粉末"
    return type_of_sample
    
```

```python
df['样品类型']=df['课题名称'].apply(sample_type)
```
