```python
import base64
```

```python
ssr_nodes=[]
with open('v2ray.txt','r') as f:
    while 1:
        ssr_url=f.readline()
        if len(ssr_url)==0:
            break
        ssr_nodes.append(ssr_url)
```

```python
ssr_nodes
```

```python
len(ssr_nodes)
```

```python
ssr_strs=[node.split('://')[-1][:-1] for node in ssr_nodes]
ssr_strs
```

```python
ssr_strs_decode=[]
index=0
for ssr_str in ssr_strs:
    ssr_str=ssr_str.replace('-','+')
    ssr_str=ssr_str.replace('_','/')
    while 1:
        try:
            ssr_str_decode=bytes.decode(base64.b64decode(bytes(ssr_str,encoding='utf-8')))
            break
        except BaseException:
            print('第'+str(index)+'条出现问题')
            ssr_str=ssr_str+'='
    index+=1
    ssr_strs_decode.append(ssr_str_decode)
```

```python
ssr_strs_decode
```

```python
ssr_strs_decode[0].split('\r\n')
```

```python
ssr_list_refine=[]
index_list=[]
index=0
for ssr in ssr_strs_decode:
    if ssr not in ssr_list_refine:
        ssr_list_refine.append(ssr)
        index_list.append(index)
    index+=1
```

```python
len(ssr_list_refine)
```

```python
index_list
```

```python
ssr_nodes_out=[ssr_nodes[i] for i in index_list]
ssr_nodes_out
```

```python
len(ssr_nodes_out)
```

```python
with open('v2ray_nodes_out.txt','w') as f:
    for ssr_node_out in ssr_nodes_out:
        f.write(ssr_node_out)
    
```
