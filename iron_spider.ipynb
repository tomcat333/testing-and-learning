```python
def wait_and_try(func):    #这个函数用来防止所需元素未加载完毕
    while 1:       
        try:
            func()
        except BaseException:
            print('wocao')
            sleep(0.1)
        else:
            print('success')
            break
```

```python
def get_method_and_sample_msg(id_num):            #已审核课题查询器，用来得到实验方案和样品信息
    driver.switch_to.parent_frame()
    driver.switch_to.frame(0)
    driver.find_element_by_name("projectNum").clear()
    driver.find_element_by_name("projectNum").send_keys(id_num)
    driver.find_element_by_xpath(u"(.//*[normalize-space(text()) and normalize-space(.)='已上传成果'])[1]/following::span[3]").click()
    def click_chakan():
        driver.find_element_by_link_text('查看').click()
    wait_and_try(click_chakan)         ####等待网页加载完毕并打开网页,然后点击查看
    driver.switch_to.parent_frame()
    driver.switch_to.frame(1)
    method=driver.find_element_by_xpath(u"(.//*[normalize-space(text()) and normalize-space(.)='三、背景意义'])[1]/following::td[4]").text
    sample=driver.find_element_by_xpath(u"(.//*[normalize-space(text()) and normalize-space(.)='五、样品信息'])/following::table[1]/tbody").text
    #sample_msg=sample.splitlines()       ####变成列表，后面导入到表格中更麻烦，干脆不转换   
    driver.switch_to.parent_frame()
    driver.find_element_by_xpath("//div[@id='maintabs']/div/div[3]/ul/li[3]/a[2]").click()  #####关闭网页
    return method,sample
```

```python
def get_input_list(input_file):   #这个函数用来将输入文件转换成input_list列表
    file=open(input_file,"r",)  
    reader=csv.reader(file)
    input_list=[]
    for input_data in reader:
        input_list.append(input_data)
    file.close()
    return input_list
```

```python
def get_message_list(input_list=[]):      #这个函数用来由输入列表得到输出列表     
    method_list=[]
    sample_list=[]
    process=1
    for content in input_list:
        method,sample=get_method_and_sample_msg(content[0])
        print("已经查完"+str(process)+"条")
        process+=1
        method_list.append(method.encode('gbk', 'ignore').decode('gbk'))#,encode和decode用来防止gbk无法解码
        sample_list.append(sample.encode('gbk', 'ignore').decode('gbk'))#,encode和decode用来防止gbk无法解码
    return method_list,sample_list
```

```python
def output_to_file(file_name,list_to_add):     #这个函数用来将输出列表输出到文件
    output_file=open(file_name,"w+",newline='')  
    writer=csv.writer(output_file)
    for result in list_to_add:
        writer.writerow(result)
    output_file.close()
```

```python
a_list=get_input_list('subject2.csv')
```

```python
m_list,s_list=get_message_list(a_list)
```

```python
output_to_file('method_and_sample_2.csv',list(np.array([a_list,m_list,s_list]).T))
```
